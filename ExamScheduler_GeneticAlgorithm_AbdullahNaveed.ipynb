```python
def fileRead(studentCourses, teachers, courses, studentNames):
    #reading data from studentCourse.csv
    file1 = open('studentCourse.csv', 'r')
    Lines = file1.readlines()
    
    for line in Lines:
        curr_record = []
        temp = line.strip()
        #tokenizing data from studentCourses table
        tokens = temp.split(",")
        for token in tokens:
            curr_record.append(token)
        studentCourses.append(curr_record)

    #reading data from teachers.csv
    file1 = open('teachers.csv', 'r')
    Lines = file1.readlines()

    for line in Lines:
        if(line.strip() != ""):
            teachers.append(line.strip())

    #reading data from courses.csv
    file1 = open('courses.csv', 'r')
    Lines = file1.readlines()

    for line in Lines:
        curr_record = []
        temp = line.strip()
        #tokenizing data from courses table
        tokens = temp.split(",")
        for token in tokens:
            curr_record.append(token)
        courses.append(curr_record)

    #reading data from studentNames.csv
    file1 = open('studentNames.csv', 'r')
    Lines = file1.readlines()

    for line in Lines:
        studentNames.append(line.strip())

studentCourses = []
teachers = []
courses = []
studentNames = []

fileRead(studentCourses, teachers, courses, studentNames)

#printing data from studentCourses
#for studentCourse in studentCourses:
    #print(studentCourse)

#printing data from teachers
#for teacher in teachers:
    #print(teacher)

#printing data from courses
#for course in courses:
    #print(course)

#printing data from studentNames
#for studentName in studentNames:
    #print(studentName)
```

## SOFT CONSTRAINTS

```python
#soft constraint # 1 :- All students and teachers shall be given a break on Friday from 1-2.
def break_on_friday(timetable):
    break_cost = 0
    token = timetable[9][0].split(" ")
    second_slot_start_time = token[2]
    
    if second_slot_start_time[0] < "2":
        #print("Friday Break Time!!")
        break_cost += 1
    
    token = timetable[8][0].split(" ")
    first_slot_end_time = token[2]
    
    if first_slot_end_time[5] > "1":
        #print("Friday Break Time!!")
        break_cost += 1
        
    return break_cost 


#------for debugging-------------    
#timeTable = createTimeTable()
#print(break_on_friday(timeTable))
```

```python
#soft constraint # 2 :- A student shall not give more than 1 exam consecutively.
def consecutive_exam_for_student(timeTable):
    
    #For fitness
    clashes_cost = 0
    clashCourses = []

    changes = False   
    
    for studentCourse in studentCourses:

        count_courses = []

        for studentCourse1 in studentCourses:

            if studentCourse[1] == studentCourse1[1]:
                count_courses.append(studentCourse1[2])

        for countDay, day in enumerate(timeTable):
            count = 0
            for countItem, item in enumerate(day):
                if(item):
                    for index in count_courses:
                        if item[0] == index:
                            count += 1
            
            #checking for clashes in next slot
            if(countDay+1 < len(timeTable)):
                for countItem, item in enumerate(timeTable[countDay+1]):
                    if(item):
                        for index in count_courses:
                            if item[0] == index:
                                count += 1
                        
            #check for clashes in previous slot 
            if(countDay > 0):
                for countItem, item in enumerate(timeTable[countDay-1]):
                    if(item):
                        for index in count_courses:
                            if item[0] == index:
                                count += 1                            

            if(count > 1):
                changes = True
                clashes_cost += 1
                    
    return clashes_cost

#------for debugging-------------   
#timeTable = createTimeTable()
#print(clash_exam_for_student(timeTable))
```

```python
#soft constraint # 3 :- If a student is enrolled in a MG course and a CS course, it is preferred that their MG course exam be held before their CS course exam.

def sequence_of_exams(timetable):    
    courses_sequence = []
    sequence_cost = 0
    for rows in timetable:
        for record in rows:
            if(record):
                if(len(record[0]) >= 4):
                    courses_sequence.append(record[0])
                
    for i in range(0,len(courses_sequence)):
        if(courses_sequence[i][0] == 'C' and courses_sequence[i][1] == 'S'): #CS course in found 
            #print("CS course")
            for j in range(i,len(courses_sequence)):#check for MG courses from THIS POINT FORWARD
                if(courses_sequence[j][0] == 'M' and courses_sequence[j][1] == 'G'):#MG course after CS Course found
                    #print("MG course after CS Course found!!!")
                    sequence_cost += 1
    
    return sequence_cost

#------for debugging-------------  
#timeTable = createTimeTable()
#print(sequence_of_exams(timeTable))
```

### Fitness Function

```python
def calculate_fitness(timeTable):    
    hard_value = 0
    
    #Count missing courses
    CountMissingCourses, missingCourses = missing_course_in_timetable(timeTable)
    
    #Count exam clashes
    CountClashesExams = clash_exam_for_student(timeTable)
    
    #Count teacher consecutive duties
    CountClashesTeachers, clashTeachers = consective_teacher_in_timetable(timeTable)
    
    #Check if enrollment is correct
    studentEnrollCheck = each_student_enrolled_correctly(timeTable)
    if(studentEnrollCheck == False):
        hard_value += 1
    
    soft_value = break_on_friday(timeTable) + consecutive_exam_for_student(timeTable) + sequence_of_exams(timeTable)
    hard_value += CountMissingCourses + CountClashesExams + CountClashesTeachers
    
    return soft_value, hard_value
```

### Crossover

```python
def crossover(timeTable):
    
    chromosomes = []
    
    totalCrossovers = 0
    for row in timeTable:
        for item in row:
            if(item):
                totalCrossovers += 1
    totalCrossovers -= 10
    
    while(totalCrossovers):
    
        #tempChromosome
        tempChromosome = []
        for row in timeTable:
            newRow = []
            for item in row:
                newRow.append(item)
            
            tempChromosome.append(newRow)

        #Get Location 1
        randomLocation1 = random.randint(0, len(tempChromosome)-1)
        randomSlot1 = random.randint(1, len(tempChromosome[randomLocation1])-1)
        while(tempChromosome[randomLocation1][randomSlot1] == None):
            randomLocation1 = random.randint(0, len(tempChromosome)-1)
            randomSlot1 = random.randint(1, len(tempChromosome[randomLocation1])-1)


        #Get Location 2
        randomLocation2 = random.randint(0, len(tempChromosome)-1)
        randomSlot2 = random.randint(1, len(tempChromosome[randomLocation2])-1)
        while(randomLocation1 == randomLocation2 and randomSlot1 == randomSlot2):
            randomLocation2 = random.randint(0, len(tempChromosome)-1)
            randomSlot2 = random.randint(1, len(tempChromosome[randomLocation2])-1)

        #Interchanging bit
        temp = tempChromosome[randomLocation1][randomSlot1]
        tempChromosome[randomLocation1][randomSlot1] = tempChromosome[randomLocation2][randomSlot2]
        tempChromosome[randomLocation2][randomSlot2] = temp

        #Add new chromosome
        chromosomes.append(tempChromosome)
        
        totalCrossovers -= 1
    
    return chromosomes
    
```

### Mutation

```python
def mutation(timeTable):
    
    chromosomes = []
    
    totalMutations = 0
    for row in timeTable:
        for item in row:
            if(item):
                totalMutations += 1
    totalMutations -= 10
    
    while(totalMutations):
    
        #tempChromosome
        tempChromosome = []
        for row in timeTable:
            newRow = []
            for item in row:
                newRow.append(item)
            
            tempChromosome.append(newRow)

        #Get Random Location
        randomLocation1 = random.randint(0, len(tempChromosome)-1)
        randomSlot1 = random.randint(1, len(tempChromosome[randomLocation1])-1)
        #while(tempChromosome[randomLocation1][randomSlot1] == None):
            #randomLocation1 = random.randint(0, len(tempChromosome)-1)
            #randomSlot1 = random.randint(1, len(tempChromosome[randomLocation1])-1)
            
        #Check if we are mutating on basis of course or teacher
        mutationBasis = random.randint(0, 1)
        
        #Changing Course
        if(mutationBasis):
            
            if(tempChromosome[randomLocation1][randomSlot1]):
                currentCourse = tempChromosome[randomLocation1][randomSlot1][0]
                currentTeacher = tempChromosome[randomLocation1][randomSlot1][1]
                
                newCourse = random.choice(courses)
                while(newCourse[0] == currentCourse):
                    newCourse = random.choice(courses)

                tempChromosome[randomLocation1][randomSlot1] = ((newCourse[0],currentTeacher))
            else:
                currentTeacher = random.choice(teachers)
                newCourse = random.choice(courses)
            
                tempChromosome[randomLocation1][randomSlot1] = ((newCourse[0],currentTeacher))

        #Changing Teacher
        else:
            if(tempChromosome[randomLocation1][randomSlot1]):
                currentTeacher = tempChromosome[randomLocation1][randomSlot1][1]
                currentCourse = tempChromosome[randomLocation1][randomSlot1][0]
                
                newTeacher = random.choice(teachers)
                while(newTeacher == currentTeacher):
                    newTeacher = random.choice(teachers)

                tempChromosome[randomLocation1][randomSlot1] = ((currentCourse,newTeacher))
            else:
                newTeacher = random.choice(teachers)
                currentCourse = random.choice(courses)
                
                tempChromosome[randomLocation1][randomSlot1] = ((currentCourse[0],newTeacher))
            
            

        #Add new chromosome
        chromosomes.append(tempChromosome)
        
        totalMutations -= 1
    
    return chromosomes
```

```python
def genectic_algorithm(timeTable):
    print_timetable(timeTable)
    soft_value,hard_value  = calculate_fitness(timeTable)
    initialFitnessValue = soft_value + hard_value
    targetFitnessValue  = initialFitnessValue/2
    
    currentFitnessValue = initialFitnessValue
    #currentFitnessValue = []
    
    #currentFitnessValue.append(initialFitnessValue)
    #currentFitnessValue.append(soft_value)
    #currentFitnessValue.append(hard_value)
    
    while(currentFitnessValue > 1000):
        
        #print("Old Fitness Value: ", currentFitnessValue[0], "(W/ Hard constaints = ", currentFitnessValue[2],
        #      " W/ Soft contraints = ",currentFitnessValue[1],")")
        #print("Applying Crossover/Mutation...")
        
        print("Old Fitness Value: ", currentFitnessValue)
        print("Applying Crossover/Mutation...")
        
        crossoverValues = []
        crossoveredChromosomes = crossover(timeTable)
        for chromosome in crossoveredChromosomes:
            soft_value,hard_value = calculate_fitness(chromosome)
            fitnessValue = soft_value + hard_value
            crossoverValues.append(fitnessValue)
            
        mutationValues = []
        mutatedChromosomes = mutation(timeTable)
        for chromosome in mutatedChromosomes:
            soft_value,hard_value = calculate_fitness(chromosome)
            fitnessValue = soft_value + hard_value
            mutationValues.append(fitnessValue)
    
        minCrossOver = min(crossoverValues)
        minMutation = min(mutationValues)
        
        if(minCrossOver < minMutation):
            if(minCrossOver <= currentFitnessValue):
                index = crossoverValues.index(minCrossOver)
                tempTable = crossoveredChromosomes[index]

                currentFitnessValue = minCrossOver

                timeTable.clear()
                for row in tempTable:
                    newRow = []
                    for item in row:
                        newRow.append(item)

                    timeTable.append(newRow)
            
        else:
            if(minMutation <= currentFitnessValue):
                index = mutationValues.index(minMutation)
                tempTable = mutatedChromosomes[index]

                currentFitnessValue = minMutation

                timeTable.clear()
                for row in tempTable:
                    newRow = []
                    for item in row:
                        newRow.append(item)

                    timeTable.append(newRow)
                
        print("New Fitness Value: ", currentFitnessValue, "\n")
        print_timetable(timeTable)

    return timeTable
```
